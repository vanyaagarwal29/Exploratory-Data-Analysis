```python
#Missing values
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
features_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>=1]
```

```python
features_with_nan
```

```python
#percentage of missing values
for feature in features_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean()*100,4), "% missing values")
```

```python
##let's plot
for feature in features_with_nan:
    data=dataset.copy()
    #let's make a variable that indicate 1 if the observations was missing otherwise 0
    data[feature]=np.where(data[feature].isnull(),1,0)   #we done this to knoe the realtionship between missing values and saleprice.
    
    #Let's calculate the mean saleprice where the info is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()  #to avoid outliers we used median
    plt.title(feature)
    plt.show()
    
```

Here with the relation btw the missing values and the dependent variiable is clearly visible. so we need to replace this missing valu with something meaningful which we will do in the feature selction engineering selection.

## Numerical variables

```python
dataset.info()
```

```python
##list of numerical variables
dataset['SalePrice'].dtypes!='object'
```

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype!='object']
```

```python
print(len(numerical_features))
dataset[numerical_features].head()
```

```python
## Temporal variables(datetime variable)
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python
##only the unique values will be shown of that feature
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
## Here we will compare the difference btw all years faetures with saleprice
data=dataset.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

Observations:

Within all the yearwise info, it is clearly visible that the new houses are more costlier than the old ones and it looks like a pareto or power law distribution.


```python
##numerical variables- 2 types
##1continuous and discrete

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<=25]
```

```python
print(len(discrete_feature))
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
##Relationship btw the discrete and saleprice
data=dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()  #to avoid outliers we used median
    plt.title(feature)
    plt.show()
    
```

```python
##continuous variable
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature]
```

```python
continuous_feature
print(len(continuous_feature))
```

```python
data=dataset.copy()
for feature in continuous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()  #to avoid outliers we used median
    plt.title(feature)
    plt.show()
```

```python
data=dataset.copy()
for feature in continuous_feature:
    data[feature].hist(bins=25)  #to avoid outliers we used median
    plt.title(feature)
    plt.show()
```

## EDA PART 2

```python
## We will be using logarithmix transformation
data=dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#outlier
data=dataset.copy()
for feature in continuous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.show()
```
