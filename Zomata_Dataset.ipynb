```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
#Here columns of df_country(Country code) will be merged with df columnn(country code) 
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
##To check dtype
final_df.dtypes
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
##Pie Chart-- Top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observations:
- Zomato's maximum records or transcations are from India after that USA and theN United Kingdoms

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations:
- When Rating is between 4.5 to 4.9-->Excellent
- When Rating is between 4.0 to 4.4-->Very Good
- When Rating is between 3.5 to 3.9--> Good
- When Rating is between 2.5 to 3.4--> Average
- When Rating is between 1.8 to 2.4-->Poor
- When Rating is 0.0-->the person has not rated.

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue= 'Rating color' ,data=ratings, palette=['white','red','orange','yellow','green','green'])
```

Observations:
- Not rated count is very high,
- Maximum no of ratings are between 2.5 to 3.4

```python
##Count plot-->This is base on rating count
sns.countplot(x="Rating color", data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
##Find the counties name that has given 0 rating
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

```python
#find out which currency is used by which country?
final_df.groupby(['Country','Currency']).size().reset_index()
```
